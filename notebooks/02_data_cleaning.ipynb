# 02 — Data Cleaning

## Objective
Address all data quality issues identified in the exploration phase 
and produce a clean dataset ready for RFM analysis.

## Contents
1. Remove duplicate rows
2. Drop missing CustomerIDs
3. Filter out cancelled invoices
4. Remove zero and negative prices
5. Save cleaned dataset

---

```python
import pandas as pd

df= pd.read_excel('../data/raw/online_retail_II.xlsx', sheet_name = 'Year 2010-2011')
```

```python
df.isnull().sum()
```

```python
df_clean = df.copy()
df_clean = df_clean.drop_duplicates()
print(f"Duplicates removed: {df.shape[0] - df_clean.shape[0]}")
```

```python
before_rows = df_clean.shape[0]
df_clean = df_clean.dropna(subset=['Customer ID'])
print(f"Null values removed: {before_rows - df_clean.shape[0]}")
```

```python
df_clean.isna().sum()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.columns
```

```python
before_rows = df_clean.shape[0]
df_clean = df_clean[df_clean['Invoice'].str.startswith('C').fillna(False)==False]
print(f"Rows with cancelled invoices removed: {before_rows - df_clean.shape[0]}")
```

```python
before_rows = df_clean.shape[0]
print(f"Zero/negative price rows: {df_clean[df_clean['Price'] <= 0].shape[0]}")
df_clean = df_clean[df_clean['Price'] > 0]
print(f"Rows removed: {before_rows - df_clean.shape[0]}")
```

```python
print(f"Dataset shape after cleaning: {df_clean.shape}")
```

## Data Quality Issues Found:

1. 135,080 missing CustomerIDs (~25% of dataset). Rows without a customerID cannot be attributed to any customer and is useless for analysis. These rows will be dropped.
2. 5268 fully duplicate rows found. These will have to be removed before doing further analysis.
3. Cancelled Invoices are prefixed with 'C' in the Invoice column to represent returns and carry negative Quantity and Price values. These have to be filtered out entirely.
4. 40 rows with zero or negative price values exist outside of cancellations. These produce meaninless Monetary calculations and will be removed.
